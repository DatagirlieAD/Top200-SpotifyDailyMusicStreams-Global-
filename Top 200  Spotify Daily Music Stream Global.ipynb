```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import random 
import bisect
import collections

#Store the url in a variable container
url = 'https://kworb.net/spotify/country/global_daily.html'

#To access the webpage content
response = requests.get(url, headers = {'User-Agent' : 'Mozilla/5.0'})

#Extracting HTML from response
page_content = response.content

#Parsing the HTML
soup = BeautifulSoup(page_content, 'lxml')

#Finding the table in the HTML
chart_table = soup.find_all('table' , attrs = {"class":"sortable"})

#initialize empty dictionary
data = []

#looping through each row and table
for table in chart_table:
    #Parse the rows in the table
    table_rows = table.find_all('tr')

    #Parsing the table header
    table_header = table.find_all('th')

    #Parsing columns from table header
    cols = []
    for item in table_header:
        if len(item.text) > 0:
            cols.append(item.text)

  
    for row in table_rows:
        table_data = row.find_all('td')
        row_data = [item.text.replace('\xa0', '').strip() if item.text.strip() else None for item in table_data]

        if len(row_data) > 0:
            data.append(row_data)

    #print(data)
    #print(cols)

    df = pd.DataFrame(data, columns = cols)

    print(df)

df.to_csv('Top 200 Spotify Daily Music Streams(Global).csv')
                        
```
